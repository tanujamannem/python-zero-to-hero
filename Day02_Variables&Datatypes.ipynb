# Day 2 â€” Variables & Data Types
Variables:
- Store data in memory
- Rules: start with letter/underscore, no spaces, case-sensitive, meaningful names

Data Types:
- int: 10
- float: 10.5
- str: "Hello"
- bool: True/False
- complex: 2+3j

Mutable vs Immutable:
- Mutable: list, dict, set
- Immutable: int, float, str, tuple

Type Casting:
- int(), float(), str(), bool()

Multiple Assignment:
- a, b, c = 1, 2, 3
- x = y = z = 5

Constants:
- No true constants; use uppercase by convention, e.g., PI = 3.14

# EXAMPLES

```python
# Creating variables
a = 10
b = 20
name = "Tanuja"
print(a, b, name)
```

```python
# Checking data types
x = 10
y = 5.5
z = "Hello"
flag = True
print(type(x), type(y), type(z), type(flag))
```

```python
# Type casting
x_str = "100"
x_int = int(x_str)
y_float = float("3.14")
z_str = str(250)
print(x_int, y_float, z_str)
```

```python
# Multiple assignment
a, b, c = 1, 2, 3
print(a, b, c)

x = y = z = 5
print(x, y, z)
```

```python
# Mutable vs Immutable examples
my_list = [1,2,3]
my_tuple = (4,5,6)
my_list[0] = 10
print(my_list)
print(my_tuple)
```

# INTERVIEW QUESTIONS

### Q1: What is a variable?
### A: Container to store data

### Q2: Difference between mutable and immutable types?
### A: Mutable can be changed (list, dict, set), Immutable cannot (int, float, str, tuple)

### Q3: What is type casting?
### A: Converting one data type to another using int(), float(), str(), bool()

### Q4: How to assign multiple variables in one line?
### A: a, b, c = 1, 2, 3

### Q5: How to assign same value to multiple variables?
### A: x = y = z = value

### Q6: How to check data type of a variable?
### A: Using type() or isinstance()

### Q7: Give example of constant usage in Python
### A: PI = 3.14

### Q8: What happens if you try to change a tuple element?
### A: Error, because tuples are immutable

### Q9: What are rules for variable naming?
### A: Start with letter/underscore, no spaces, case-sensitive, meaningful names

### Q10: Difference between mutable and immutable objects in memory
###A: Mutable objects can be changed in-place, immutable objects create new object on change
