```python
# ===========================
# Day 2 â€” Variables & Data Types
# ===========================

# ===========================
# THEORY
# ===========================
theory = """
Variables:
- Store data in memory
- Rules: start with letter/underscore, no spaces, case-sensitive, meaningful names

Data Types:
- int: 10
- float: 10.5
- str: "Hello"
- bool: True/False
- complex: 2+3j

Mutable vs Immutable:
- Mutable: list, dict, set
- Immutable: int, float, str, tuple

Type Casting:
- int(), float(), str(), bool()

Multiple Assignment:
- a, b, c = 1, 2, 3
- x = y = z = 5

Constants:
- No true constants; use uppercase by convention, e.g., PI = 3.14
"""
print(theory)

# ===========================
# EXAMPLES
# ===========================
# Creating variables
a = 10
b = 20
name = "Tanuja"
print(a, b, name)

# Checking data types
x = 10
y = 5.5
z = "Hello"
flag = True
print(type(x), type(y), type(z), type(flag))

# Type casting
x_str = "100"
x_int = int(x_str)
y_float = float("3.14")
z_str = str(250)
print(x_int, y_float, z_str)

# Multiple assignment
a, b, c = 1, 2, 3
print(a, b, c)

x = y = z = 5
print(x, y, z)

# Mutable vs Immutable examples
my_list = [1,2,3]
my_tuple = (4,5,6)
my_list[0] = 10
print(my_list)
print(my_tuple)

# ===========================
# PRACTICE QUESTIONS
# ===========================
# Q1: Create variables for marks of 3 subjects and print total
m1 = 85
m2 = 90
m3 = 78
total = m1 + m2 + m3
print(total)

# Q2: Convert int to float and float to int
x = 50
y = 3.75
print(float(x))
print(int(y))

# Q3: Swap two variables
a = 15
b = 25
a, b = b, a
print(a, b)

# Q4: Assign same value to multiple variables
x = y = z = 100
print(x, y, z)

# Q5: Check type of each variable
print(type(m1), type(y), type("Python"), type(True))

# ===========================
# CHALLENGE QUESTIONS
# ===========================
# Challenge 1: Create complex number and print real & imaginary parts
c = 3 + 4j
print(c.real, c.imag)

# Challenge 2: Calculate area of rectangle (length*width)
length = 10
width = 5
area = length * width
print(area)

# Challenge 3: Check if a variable is int
x = 20
print(isinstance(x, int))

# Challenge 4: Multiple assignment sum
a, b, c = 5, 10, 15
print(a + b + c)

# Challenge 5: Constant usage example
PI = 3.14
radius = 7
circumference = 2 * PI * radius
print(circumference)

# ===========================
# INTERVIEW QUESTIONS
# ===========================
# Q1: What is a variable?
# A: Container to store data

# Q2: Difference between mutable and immutable types?
# A: Mutable can be changed (list, dict, set), Immutable cannot (int, float, str, tuple)

# Q3: What is type casting?
# A: Converting one data type to another using int(), float(), str(), bool()

# Q4: Practical: Sum of 3 numbers
num1 = 10
num2 = 20
num3 = 30
print(num1 + num2 + num3)

# Q5: Practical: Convert float to int and add
f = 7.8
i = int(f)
print(i + 5)
```
