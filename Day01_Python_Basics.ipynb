# Day 1 — Python Basics

Python:
- High-level, interpreted, dynamically typed language
- Uses indentation instead of braces for code blocks

Variables:
- Store data in memory
- Rules: start with letter/underscore, no spaces, case-sensitive

Data Types:
- int: 10
- float: 10.5
- str: "Hello"
- bool: True/False

Operators:
- Arithmetic: +, -, *, /, %, **, //
- Comparison: ==, !=, >, <, >=, <=
- Logical: and, or, not

Comments:
- Single line: #
- Multi-line: '''...''' or """ """

Print & Type:
- print() to display output
- type() to check data type


# EXAMPLES

```python
# Variable assignment
a = 10
b = 20.5
name = "Tanuja"
is_active = True
print(a, b, name, is_active)
```

```python
# Operators
print(a + b)       # Addition
print(a ** 2)      # Exponent
print(a > b)       # Comparison
print(a == 10 and b > 10)  # Logical
```
