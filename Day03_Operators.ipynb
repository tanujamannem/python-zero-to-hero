```python
# ===========================
# Day 3 â€” Operators
# ===========================

# ===========================
# THEORY
# ===========================
theory = """
Operators in Python:

1. Arithmetic Operators:
+  Addition
-  Subtraction
*  Multiplication
/  Division
%  Modulus (remainder)
** Exponent
// Floor division

2. Comparison Operators:
== Equal
!= Not equal
>  Greater than
<  Less than
>= Greater than or equal
<= Less than or equal

3. Assignment Operators:
=   Assign
+=  Add and assign
-=  Subtract and assign
*=  Multiply and assign
/=  Divide and assign
%=  Modulus and assign

4. Logical Operators:
and  Logical AND
or   Logical OR
not  Logical NOT

5. Bitwise Operators:
&  AND
|  OR
^  XOR
~  NOT
<< Left shift
>> Right shift
"""
print(theory)

# ===========================
# EXAMPLES
# ===========================
# Arithmetic
a = 10
b = 3
print(a + b)
print(a - b)
print(a * b)
print(a / b)
print(a % b)
print(a ** b)
print(a // b)

# Comparison
x = 10
y = 20
print(x == y)
print(x != y)
print(x > y)
print(x < y)
print(x >= y)
print(x <= y)

# Assignment
c = 5
c += 3
print(c)
c -= 2
print(c)
c *= 2
print(c)
c /= 2
print(c)
c %= 3
print(c)

# Logical
p = True
q = False
print(p and q)
print(p or q)
print(not p)

# Bitwise
x = 5   # 0101
y = 3   # 0011
print(x & y)
print(x | y)
print(x ^ y)
print(~x)
print(x << 1)
print(x >> 1)

# ===========================
# PRACTICE QUESTIONS
# ===========================
# Q1: Add, subtract, multiply two numbers
a = 7
b = 4
print(a + b)
print(a - b)
print(a * b)

# Q2: Division, modulus, exponent
print(a / b)
print(a % b)
print(a ** b)

# Q3: Compare two numbers
print(a == b)
print(a != b)
print(a > b)
print(a < b)

# Q4: Logical AND, OR, NOT
x = True
y = False
print(x and y)
print(x or y)
print(not x)

# Q5: Assignment operators
c = 10
c += 5
c *= 2
print(c)

# ===========================
# CHALLENGE QUESTIONS
# ===========================
# Challenge 1: Bitwise AND, OR, XOR
a = 12  # 1100
b = 5   # 0101
print(a & b)
print(a | b)
print(a ^ b)

# Challenge 2: Left and right shift
x = 7
print(x << 2)
print(x >> 1)

# Challenge 3: Combined arithmetic
a = 5
b = 3
c = 2
result = a + b * c - b / a
print(result)

# Challenge 4: Logical expression
x = True
y = False
z = True
print((x and y) or (not z))

# Challenge 5: Using assignment operators
num = 10
num += 5
num *= 3
num -= 4
num //= 2
print(num)

# ===========================
# INTERVIEW QUESTIONS
# ===========================
# Q1: Name arithmetic operators in Python
# A: +, -, *, /, %, **, //

# Q2: Difference between / and //?
# A: / returns float division, // returns floor division

# Q3: Logical operators and their use
# A: and, or, not; used to combine boolean expressions

# Q4: Practical: Evaluate expression
x = 10
y = 5
z = 2
print((x + y) * z - y)

# Q5: Practical: Bitwise operation
a = 9
b = 4
print(a & b)
print(a | b)
print(a ^ b)
```
