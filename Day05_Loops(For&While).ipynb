# Day 5 â€” Loops (for & while)

Loops in Python:
- Used to execute a block of code multiple times

Types:
1. for loop:
   - Iterates over a sequence (list, tuple, string, range)
   Syntax:
   for variable in sequence:

       code block

2. while loop:
   - Executes as long as condition is True
   Syntax:
   while condition:
   
       code block

Loop Control Statements:
- break: exits the loop immediately
- continue: skips current iteration
- pass: does nothing (placeholder)

Nested Loops:
- Loop inside another loop

range() function:
- Generates sequence of numbers
- range(stop)
- range(start, stop)
- range(start, stop, step)

Common Uses:
- Iterating sequences
- Repeating tasks
- Generating patterns


## EXAMPLES

```python
# For loop
fruits = ["apple", "banana", "cherry"]
for fruit in fruits:
    print(fruit)
```

```python
# Using range
for i in range(5):
    print(i)  # 0 to 4

for i in range(1, 6):
    print(i)  # 1 to 5

for i in range(0, 10, 2):
    print(i)  # 0,2,4,6,8
```

```python
# While loop
count = 0
while count < 5:
    print(count)
    count += 1
```

```python
# Break statement
for i in range(10):
    if i == 5:
        break
    print(i)
```

```python
# Continue statement
for i in range(5):
    if i == 3:
        continue
    print(i)
```

```python
# Nested loops (pattern)
for i in range(1, 4):
    for j in range(1, i+1):
        print(j, end=" ")
    print()
```

```python
# Q2: Print even numbers between 1 and 20
for i in range(2, 21, 2):
    print(i)
```

```python
# Q3: Print numbers 10 to 1 using for loop
for i in range(10, 0, -1):
    print(i)
```

```python
# Q4: Print numbers 1 to 5 using while loop
n = 1
while n <= 5:
    print(n)
    n += 1
```

## CHALLENGE QUESTIONS

```python
# Challenge 1: Print numbers divisible by 3 from 1 to 30
for i in range(1, 31):
    if i % 3 == 0:
        print(i)
```

```python
# Challenge 2: Print first 10 Fibonacci numbers
a, b = 0, 1
for _ in range(10):
    print(a, end=' ')
    a, b = b, a+b
print()
```

```python
# Challenge 3: Print square of numbers 1 to 10
for i in range(1, 11):
    print(i**2)
```

```python
# Challenge 4: Print reverse pattern
for i in range(5, 0, -1):
    print('*' * i)
```

```python
# Challenge 5: Count number of vowels in a string
s = "Python is fun"
count = 0
for ch in s.lower():
    if ch in 'aeiou':
        count += 1
print("Vowels:", count)
```

```python
# Challenge 6: Print prime numbers between 1 and 20
for num in range(2, 21):
    for i in range(2, num):
        if num % i == 0:
            break
    else:
        print(num, end=' ')
print()
```

```python
# Challenge 7: Sum of all even numbers in a list
lst = [1,2,3,4,5,6]
sum_even = 0
for i in lst:
    if i % 2 == 0:
        sum_even += i
print(sum_even)
```

```python
# Challenge 8: Print multiplication table in reverse
for i in range(10, 0, -1):
    print('7 x', i, '=', 7*i)
```

```python
# Challenge 9: Print triangle pattern with numbers
for i in range(1, 6):
    for j in range(1, i+1):
        print(j, end='')
    print()
```

```python
# Challenge 10: Using while loop, find factorial of 5
n = 5
fact = 1
while n > 0:
    fact *= n
    n -= 1
print(fact)
```


## INTERVIEW QUESTIONS

#### Q1: What is a loop?

#### Q3: What is the purpose of break and continue?
#### A: break exits loop, continue skips current iteration

#### Q4: Can loops be nested?
#### A: Yes, one loop inside another

####Q8: How to generate a sequence of numbers?
