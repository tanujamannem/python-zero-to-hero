# Day 4 â€” Conditional Statements

Conditional Statements:
- Used to make decisions based on conditions
- Executes code only if condition is True

if statement:
- Executes block if condition is True
Syntax:
if condition:

     code

if-else statement:
- Executes one block if condition is True, else another block
Syntax:
if condition:

     code if True
else:

   code if False

if-elif-else statement:
- Used for multiple conditions
Syntax:
if condition1:

     code1
elif condition2:
     code2
else:
     code3

Nested Conditions:
- Conditions inside another condition

Logical Operators:
- and: True if both True
- or: True if any True
- not: True if False, False if True

Comparison Operators:
- ==, !=, >, <, >=, <=

Truthiness:
- Non-zero numbers and non-empty sequences are True
- Zero, None, empty sequences are False

# EXAMPLES

```python
# Simple if
x = 10
if x > 5:
    print("x is greater than 5")
```

```python
# if-else
y = 3
if y % 2 == 0:
    print("Even")
else:
    print("Odd")
```

```python
# if-elif-else
marks = 85
if marks >= 90:
    print("Grade A")
elif marks >= 75:
    print("Grade B")
elif marks >= 50:
    print("Grade C")
else:
    print("Fail")
```

```python
# Nested if
num = 15
if num > 0:
    if num % 2 == 0:
        print("Positive Even")
    else:
        print("Positive Odd")
```

```python
# Logical operators
a = 10
b = 20
if a > 5 and b > 15:
    print("Both conditions True")

if a < 5 or b > 15:
    print("At least one condition True")

if not a < 5:
    print("Not a < 5")
```

# PRACTICE QUESTIONS

```python
# Q1: Check if a number is positive
n = 7
if n > 0:
    print("Positive")
```

```python
# Q2: Check if a number is even or odd
n = 8
if n % 2 == 0:
    print("Even")
else:
    print("Odd")
```

```python
# Q3: Check grade based on marks
marks = 72
if marks >= 90:
    print("Grade A")
elif marks >= 75:
    print("Grade B")
elif marks >= 50:
    print("Grade C")
else:
    print("Fail")
```

```python
# Q4: Nested if - positive even/odd
num = 12
if num > 0:
    if num % 2 == 0:
        print("Positive Even")
    else:
        print("Positive Odd")
else:
    print("Non-positive")
```

```python
# Q5: Logical operators - check range
x = 15
if x > 10 and x < 20:
    print("x is between 10 and 20")
```

```python
# Q6: Check if character is vowel
ch = 'e'
if ch in 'aeiouAEIOU':
    print("Vowel")
else:
    print("Not vowel")
```

```python
# Q7: Check if year is leap year
year = 2024
if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
    print("Leap year")
else:
    print("Not leap year")
```

```python
# Q8: Check if string is empty
s = ""
if s:
    print("String has content")
else:
    print("Empty string")
```

```python
# Q9: Check number sign
num = -5
if num > 0:
    print("Positive")
elif num < 0:
    print("Negative")
else:
    print("Zero")
```

```python
# Q10: Multiple conditions with logical operators
a, b, c = 5, 10, 15
if a < b and b < c:
    print("b is between a and c")
```

# CHALLENGE QUESTIONS

```python
# Challenge 1: Check if number is divisible by 3 and 5
num = 30
if num % 3 == 0 and num % 5 == 0:
    print("Divisible by 3 and 5")
```

```python
# Challenge 2: Find largest of 3 numbers
a, b, c = 10, 25, 15
if a > b and a > c:
    print("a is largest")
elif b > c:
    print("b is largest")
else:
    print("c is largest")
```

```python
# Challenge 3: Check if character is uppercase or lowercase
ch = 'G'
if ch.isupper():
    print("Uppercase")
else:
    print("Lowercase")
```

```python
# Challenge 4: Nested logical conditions
x, y = 7, 20
if x < 10:
    if y > 15:
        print("x<10 and y>15")
```

```python
# Challenge 5: Check if number is within range 50-100
n = 75
if 50 <= n <= 100:
    print("Within range")
```

```python
# Challenge 6: Password validation (min 6 chars, contains digit)
password = "abc123"
if len(password) >= 6 and any(c.isdigit() for c in password):
    print("Valid password")
```

```python
# Challenge 7: Print day of week using number
day = 3
if day == 1:
    print("Monday")
elif day == 2:
    print("Tuesday")
elif day == 3:
    print("Wednesday")
elif day == 4:
    print("Thursday")
elif day == 5:
    print("Friday")
elif day == 6:
    print("Saturday")
else:
    print("Sunday")
```

```python
# Challenge 8: Check if number is prime (basic)
n = 7
if n > 1:
    for i in range(2, n):
        if n % i == 0:
            print("Not prime")
            break
    else:
        print("Prime")
```

```python
# Challenge 9: Check if list has any positive numbers
lst = [-1, -5, 10, -2]
if any(x > 0 for x in lst):
    print("Has positive numbers")
```

```python
# Challenge 10: Check if all elements in list are even
lst = [2, 4, 6]
if all(x % 2 == 0 for x in lst):
    print("All even")
```
